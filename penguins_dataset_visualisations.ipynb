```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
penguins=sns.load_dataset('penguins')
penguins
```

```python
penguins['species'].value_counts()
```

```python
sns.countplot(data=penguins,x='species')
plt.show()
```

```python
penguins['sex'].value_counts()
```

```python
sns.countplot(data=penguins,x='sex')
plt.show()
```

```python
penguins['island'].value_counts()
```

```python
sns.countplot(data=penguins,x='island')
```

```python
penguins=penguins.dropna()
penguins
```

```python
sns.scatterplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='sex')
plt.show()
```

```python
sns.lineplot(data=penguins, x="bill_length_mm", y="body_mass_g")
```

```python
sns.barplot(data=penguins,x="species",y="body_mass_g")
```

```python
sns.histplot(data=penguins,x='body_mass_g',bins=20)
plt.show()
```

```python
sns.kdeplot(data=penguins,x='body_mass_g',fill=True)
plt.show()
```

```python
sns.boxplot(data=penguins,x='species',y='body_mass_g')
plt.show()
```

```python
sns.displot(data=penguins,x='flipper_length_mm',kde=True)
plt.show()
```

```python
sns.jointplot(data=penguins,x='bill_length_mm',y='body_mass_g',kind='scatter')
plt.show()
```

```python
sns.violinplot(data=penguins, y="body_mass_g")
#Since many penguins weigh around 3200 g, the violin becomes wider there.
```

```python
sns.pairplot(data=penguins,hue='species')
```

```python
corr=penguins.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr,annot=True,cmap='Spectral')
```

```python
penguins.groupby(["species","sex"]).mean(numeric_only=True)
```

```python
pd.crosstab(
    penguins["species"],
    penguins["sex"]
)
```

```python
penguins.groupby(['species','island']).mean(numeric_only=True)
```

```python
avg_bodymass=penguins['body_mass_g'].mean()
print("Average body mass of penguins(g): ",avg_bodymass)

median_bodymass=penguins['body_mass_g'].median()
print("Median body mass of penguins: ",median_bodymass)

mode_bodymass=penguins['body_mass_g'].std()
print("Mode body mass of penguins: ",mode_bodymass)

print(f"Highest Body mass: {penguins['body_mass_g'].max()}")
```
