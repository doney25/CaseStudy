1. Bank Account Manager<br>
a. Simulate a bank account with deposit, withdrawal, and balance
enquiry operations<br>
b. Use variables and data types to store account information<br>
c. Implement conditional statements to check for sufficient balance
before withdrawal<br>
d. Use functions to perform deposit, withdrawal, and balance inquiry
operations.<br>

```python

name=input("Enter your name: ")
acc_no=int(input("Enter your account number: "))
balance=int(input("Enter your balance: "))
def deposit(amount):
  global balance
  balance+=amount
  print(f"Balance after depositing {amount}: {balance}")

def withdraw(amount):
  global balance
  if amount<=balance:
    balance-=amount
    print(f"Balance after withdrawing {amount}: {balance} ")
  else:
    print("Insufficient balance")


deposit_amount=int(input("Amount you wish to deposit: "))
deposit(deposit_amount)
withdraw_amount=int(input("Amount you wish to withdraw: "))
withdraw(withdraw_amount)
```
